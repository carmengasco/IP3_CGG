5.	Remove all the symbols (\$\,)from the 'Price' column and convert  and update it to  64-bit int numbers using astype(‘int64’) Hint. Look for string functions like replace/strip). The data frame will be look like after that:

```python
#Removing the $ symbol
car_sales['Price'] = car_sales['Price'].replace({'\$':''}, regex=True)
car_sales.head()
```

```python
#Converting into int64
#We also need to replace ',' with ''
car_sales['Price'] = car_sales['Price'].replace({',':''}, regex=True)
car_sales['Price']=car_sales['Price'].astype('int64')
car_sales.info()
```

```python
car_sales.head()
```

6.	Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.

 
16.	Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation 


```python
heat_map=car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr(method='pearson')
```

```python
axis=sns.heatmap(heat_map, annot=True, cmap='Blues')
```

The correlation value tells us how much are two variables related. The positive coefficient states that if one increases, the other one is also going to increase. A negative one states that if one variable is increasing, the other one is going to decrease. The diaganonal is 1 because each variable is 100% correlated with itself. We can also notice that it is symmetrical. This means that the correlation of x with y is the same as the one of y with x.
